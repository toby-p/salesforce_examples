# Salesforce Formula
## Convert Pence/Cents Value from a Currency Field to String
Sometimes it is necessary to have currency values written in full when generating contract documents. For example when generating a rent agreement, £2500 per month would be written in the contract as:

> "TWO THOUSAND FIVE HUNDRED POUNDS PER MONTH"

The formula below can be used to help with this, by generating the value after the decimal point (i.e. the pence/cents value) in a written format. For example, if the currency field value was £29.27, this formula would return "TWENTY SEVEN PENCE".
